## Encode music21 stream to text representation and back

### Load midi data

```python
path = Path('../data/midi')
hook_path = path/'hooktheory'
hook_metapath = hook_path/'song_metadata.json'
metadata = load_json(hook_metapath)
song_info = list(metadata.values())[800]; song_info
```

```python
midi_file = song_info['midi_keyc']
```

### Encode/Decode

```python
string_repr = midi2str(midi_file); string_repr[:100]
```

```python
out_stream = str2stream(string_repr)
```

### Step by step encoding

```python
stream = file2stream(song_info['midi_keyc'])
```

```python
s_arr = stream2nparr(stream); s_arr.sum()
```

```python
seq = nparr2seq(s_arr)
```

```python
seq_str = seq2str(seq)
```

### Decoding

```python
res = str2seq(seq_str)
```

```python
np_arr = seq2numpy(res); np_arr.sum()
```

```python
s = arr2stream(np_arr)
```
